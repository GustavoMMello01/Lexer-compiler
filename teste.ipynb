```python
import re
import math

T_KEYWORD = "keyword"
T_OP = "op"
T_FLOAT = "float"
T_STRING = "string"
T_ID = "id"
T_EOF = "eof"
T_DELIMITER = "delimiter"
T_BLOCK = "block"
T_PARA = "parenthesis"
T_VAR_TYPE = "var_type"
T_IF = "if"
T_ELIF = "elif"
T_ELSE = "else"
T_OP_COMP = 'T_OP_COMP'
T_PRINT = "print"
T_WHILE = 'while'
T_FOR = 'for'
T_ARROW_UP = "arrow_up"
T_ARROW_DOWN = "arrow_down"
T_COMMA = "comma"
T_FUNC = "func"
T_RETURN = "return"
T_MAIN = "main"

class Scope():

    def __init__(self, parent=None):
        self.symbols = {}
        self.parent = parent

    def doesSymbolExist(self, symbol):
        if symbol in self.symbols:
            return self
        if self.parent is not None:
            return self.parent.doesSymbolExist(symbol)
        return None

    def addSymbol(self, symbol):
        print("Added Symbol: ", symbol)
        self.symbols[symbol] = None

        print("Symbols: ", self.symbols)

class Token():

    def __init__(self, tipo, valor=None):
        self.tipo = tipo
        self.valor = valor

    def __str__(self):
        return '<%s %s>' % (self.tipo, self.valor)

    def __repr__(self):
        return self.__str__()


class StopExecution(Exception):
    def _render_traceback_(self):
        pass

def afd_float(token):
    try:
        float_pattern = r'^-?\d+\.\d+$'
        if re.match(float_pattern, str(token)):
            return True
        return False
    except:
        return False

def afd_string(token):
    if token[0] == '"' and token[-1] == '"':
        if '"' not in token[1:-1]:
            return True
        else:
            raise ValueError('Aspas em um local inesperado.')
    else:
        return False

def afd_identificador(token):
    regex = re.compile('[a-zA-Z0-9_]+')
    r = regex.match(token)
    if r is not None:
        if r.group() == token:
            return True
        else:
            return False
    else:
        return False

def afd_return(token):
    return token == "return"

def afd_print(token):
    return token == "print"

def afd_if(token):
    return token == "if"

def afd_elif(token):
    return token == "elif"

def afd_else(token):
    return token == "else"

def afd_delimiter(token):
    return token == ";"

def afd_block(token):
    return token in ["{", "}"]

def afd_var_type(token):
    return token in ["var"]

def afd_para(token):
    return token in ["(", ")"]

def afd_principal(token):
    if token == "main":
        return Token(T_MAIN, 'main')

    elif token in "=+-*/^":
        return Token(T_OP, token)

    elif token in ["==", "!=", "<", ">", "<=", ">="]:
        return Token(T_OP_COMP, token)

    if token == "func":
        return Token(T_FUNC, token)

    elif afd_return(token):
        return Token(T_RETURN, token)

    elif afd_float(token):
        return Token(T_FLOAT, token)

    elif afd_string(token):
        return Token(T_STRING, token)

    elif afd_if(token):
        return Token(T_IF, token)

    elif afd_elif(token):
        return Token(T_ELIF, token)

    elif afd_else(token):
        return Token(T_ELSE, token)

    if token == 'while':
        return Token(T_WHILE, token)

    elif token == 'for':
        return Token(T_FOR, token)

    elif token == "->":
        return Token(T_ARROW_UP, token)
    elif token == "<-":
        return Token(T_ARROW_DOWN, token)

    elif token == ",":
        return Token(T_COMMA, token)

    elif afd_print(token):
        return Token(T_PRINT, token)

    elif afd_delimiter(token):
        return Token(T_DELIMITER, token)

    elif afd_block(token):
        return Token(T_BLOCK, token)

    elif afd_var_type(token):
        return Token(T_VAR_TYPE, token)

    elif afd_para(token):
        return Token(T_PARA, token)

    elif afd_identificador(token):
        return Token(T_ID, token)

    else:
        raise ValueError('Valor inesperado')

class Parser():

    def __init__(self, tokens):
        self.tokens = tokens
        self.pos = -1
        self.token_atual = None
        self.currentScope = Scope()
        self.ir = []

        self.proximo()

    def generate_ir(self):
        return "\n".join(self.ir)

    def proximo(self):
        self.pos += 1

        if self.pos >= len(self.tokens):
            self.token_atual = Token(T_EOF)
        else:
            self.token_atual = self.tokens[self.pos]

        print(f"Token atual: {self.token_atual}, PosiÃ§Ã£o: {self.pos}")
        return self.token_atual


    def erro(self, expected=None, message=None):
        error_message = f'Erro de sintaxe. Esperado: {expected}, Encontrado: {self.token_atual}'
        if message:
            error_message += f" {message}"
        raise Exception(error_message)


    def use(self, tipo, valor=None):
        if self.token_atual.tipo != tipo or (valor is not None and self.token_atual.valor != valor):
            self.erro(expected=f"{tipo}{' ' + valor if valor else ''}")
        else:
            self.proximo()

    def instructions(self):
        """
        statements ::= <instruction> <instructions>
        """
        while self.token_atual.tipo in [T_VAR_TYPE, T_ID, T_IF, T_ELIF, T_ELSE, T_PRINT, T_WHILE, T_FOR]:
            self.instruction()
            if self.token_atual.tipo == T_BLOCK and self.token_atual.valor == "}":
                break


    def instruction(self):
        print("Parsing instruction...")
        if self.token_atual.tipo == T_VAR_TYPE:
            self.declare_or_assign_variable()
        elif self.token_atual.tipo in [T_IF, T_ELIF, T_ELSE]:
            self.control_structure()
        elif self.token_atual.tipo == T_FOR:
            self.parse_for()
        else:
            self.statement()
            if self.token_atual.tipo != T_BLOCK:
                self.use(T_DELIMITER, ";")

    def declare_or_assign_variable(self):
        type = self.token_atual.valor
        self.use(T_VAR_TYPE, "var")
        name = self.token_atual.valor
        self.use(T_ID)

        if self.currentScope.doesSymbolExist(name):
            raise Exception(f"Erro semantico. Simbolo {name} foi declarado mais de uma vez.")

        if self.peek().tipo == T_OP and self.peek().valor == '=':
            self.use(T_OP, '=')
            value, str_expr = self.expr()
            self.currentScope.symbols[name] = value
            self.ir.append(f"float {name} = {str_expr};")
        else:
            self.currentScope.addSymbol(name)
            self.ir.append(f"float {name};")
            self.use(T_DELIMITER, ";")

    def start(self):
        """
        start ::= { function_definition* main_block }
        """
        # Processa todas as definiÃ§Ãµes de funÃ§Ã£o
        while self.token_atual.tipo == T_FUNC:
            self.parse_function_definition()

        # Procura pelo token 'main'
        if self.token_atual.tipo != T_MAIN:
            raise Exception("Erro de sintaxe. Esperado: 'main' para o inÃ­cio do bloco principal.")

        # Processa o bloco main
        self.main_program()

        # Insere os includes no inÃ­cio do IR
        self.ir.insert(0, "#include <math.h>")
        self.ir.insert(0, "#include <stdio.h>")

    def main_program(self):
        """
        Parses the main program block.
        """
        self.use(T_MAIN)
        self.use(T_BLOCK, "{")

        self.ir.append("int main() {")

        self.instructions()

        self.ir.append("}")

        self.use(T_BLOCK, "}")

    def parse_function_definition(self):
        print("Parsing function definition...")
        self.use(T_FUNC)

        func_name = self.token_atual.valor
        self.use(T_ID)
        self.use(T_PARA, "(")
        params = self.parse_parameters()
        self.use(T_PARA, ")")

        param_placeholders = {param: 'param_placeholder' for param in params}

        self.funcionScope = Scope(self.currentScope)
        self.currentScope = self.funcionScope

        self.currentScope.symbols.update(param_placeholders)

        self.use(T_BLOCK, "{")
        current_ir = self.ir  # Guarda o IR atual
        self.ir = []  # Reseta o IR para a funÃ§Ã£o
        self.instructions()
        func_body_ir = self.ir
        self.use(T_BLOCK, "}")

        # Gerar o cÃ³digo C para a funÃ§Ã£o
        function_ir = [f"void {func_name}({', '.join(['float ' + p for p in params])}) {{"] + func_body_ir + ["}"]
        self.ir = current_ir + function_ir  # Concatena com o IR global

        #for param in params:
            #del self.currentScope.symbols[param]

        self.currentScope = self.currentScope.parent

    def parse_parameters(self):
        params = []
        while self.token_atual.tipo != T_PARA or (self.token_atual.tipo == T_PARA and self.token_atual.valor != ")"):
            if self.token_atual.tipo == T_ID:
                params.append(self.token_atual.valor)
                self.use(T_ID)
            elif self.token_atual.tipo == T_COMMA:
                self.use(T_COMMA)
            else:
                self.erro("Expected an identifier or comma in function parameters.")
        return params

    def parse_instructions(self):
        """
        Parses a sequence of instructions until a closing block token is encountered.
        """
        instructions = []
        while self.token_atual.tipo != T_BLOCK or (self.token_atual.tipo == T_BLOCK and self.token_atual.valor != "}"):
            self.statement()
            instructions.extend(self.ir)
            self.ir = []
            self.proximo()
        return instructions

    def function_call(self):
        func_name = self.token_atual.valor
        self.use(T_ID)
        self.use(T_PARA, "(")
        args = []

        while not (self.token_atual.tipo == T_PARA and self.token_atual.valor == ")"):
            if self.token_atual.tipo in [T_ID, T_FLOAT]:
                args.append(self.token_atual.valor)
                self.use(self.token_atual.tipo)
            elif self.token_atual.tipo == T_COMMA:
                self.use(T_COMMA)

        self.use(T_PARA, ")")
        args_str = ', '.join(args)
        self.ir.append(f"{func_name}({args_str});")


    def statement(self):
        """
        statement ::= <id> <op => expr | control_structure
        """

        print("Parsing statement...")
        if self.token_atual.tipo == T_IF:
            self.control_structure()
        elif self.token_atual.tipo == T_ID:
            next_token = self.peek()
            if next_token.tipo == T_PARA:
                self.function_call()
            else:
                name = self.token_atual.valor
                self.use(T_ID)
                self.use(T_OP, '=')
                value, str_expr = self.expr()
                if not self.currentScope.doesSymbolExist(name):
                    raise Exception(f"Erro semantico. Simbolo {name} nao foi declarado.")
                self.currentScope.symbols[name] = value
                self.ir.append(f"{name} = {str_expr};")
        elif self.token_atual.tipo == T_PRINT:
            self.parse_print()
        elif self.token_atual.tipo == T_WHILE:
            self.parse_while()
        elif self.token_atual.tipo == T_FOR:
            self.parse_for()
        elif self.token_atual.tipo == T_RETURN:
            self.parse_return()
        elif self.token_atual.tipo == T_VAR_TYPE:
            self.instruction()
        else:
            self.erro("InstruÃ§Ã£o inesperada.")

    def parse_return(self):
        print("Parsing return statement...")
        self.use(T_RETURN)

        _, return_expr = self.expr()

        self.use(T_DELIMITER, ";")  # Consume the semicolon

        # Adiciona o cÃ³digo C para o return
        self.ir.append(f"return {return_expr};")

    def peek(self):
        """
        Returns the next token without consuming it.
        """
        if self.pos + 1 < len(self.tokens):
            return self.tokens[self.pos + 1]
        else:
            return Token(T_EOF)

    def parse_while(self):
        print("Parsing while loop...")

        self.whileScope = Scope(self.currentScope)
        self.currentScope = self.whileScope

        self.use(T_WHILE)
        self.use(T_PARA, "(")
        condition, condition_str = self.expr()
        self.use(T_PARA, ")")
        self.use(T_BLOCK, "{")
        self.ir.append(f"while ({condition_str}) {{")
        self.instructions()
        self.use(T_BLOCK, "}")
        self.ir.append("}")

        self.currentScope = self.currentScope.parent

    def parse_for(self):
        """
        Parses a for loop.
        for ( a -> b , c ) {
            print ( a )
        }
        """
        print("Parsing for loop...")

        self.forScope = Scope(self.currentScope)
        self.currentScope = self.forScope

        self.use(T_FOR)
        self.use(T_PARA, "(")

        start_variable = self.token_atual.valor
        self.use(T_ID)

        if self.token_atual.valor == '->':
            direction = 'increment'
            self.use(T_ARROW_UP)
        elif self.token_atual.valor == '<-':
            direction = 'decrement'
            self.use(T_ARROW_DOWN)

        end_variable = self.token_atual.valor
        self.use(T_ID)
        self.use(T_COMMA)

        _, step_expression = self.expr()  # Retorna o valor da expressÃ£o e a string que representa a expressÃ£o

        self.use(T_PARA, ")")

        # Gera o cÃ³digo C para o loop for
        for_init = f"float {start_variable} = {start_variable};"
        for_cond = f"{start_variable} {'<=' if direction == 'increment' else '>='} {end_variable};"
        for_incr = f"{start_variable} {'+=' if direction == 'increment' else '-='} {step_expression}"
        self.ir.append(f"for ({for_init} {for_cond} {for_incr}) {{")

        self.use(T_BLOCK, "{")
        self.instructions()
        self.use(T_BLOCK, "}")

        self.ir.append("}")

        self.currentScope = self.currentScope.parent

    def parse_print(self):
        """
        Parses a print statement.
        """
        print("Parsing print statement...")
        self.use(T_PRINT)
        self.use(T_PARA, "(")
        format_strings = []
        print_arguments = []

        while self.token_atual.tipo != T_PARA or (self.token_atual.tipo == T_PARA and self.token_atual.valor != ")"):
            if self.token_atual.tipo == T_STRING:
                # Strings literais
                format_strings.append("%s")
                print_arguments.append(self.token_atual.valor)
                self.use(T_STRING)
            elif self.token_atual.tipo == T_ID:
                # Identificadores (variÃ¡veis)
                var_name = self.token_atual.valor
                format_strings.append("%f" if isinstance(self.currentScope.doesSymbolExist(var_name).symbols, float) else "%f")
                print_arguments.append(var_name)
                self.use(T_ID)
            elif self.token_atual.tipo == T_FLOAT:
                # Literais numÃ©ricos
                format_strings.append("%f")
                print_arguments.append(self.token_atual.valor)
                self.use(self.token_atual.tipo)

            if self.token_atual.tipo == T_COMMA:
                self.use(T_COMMA)

        self.use(T_PARA, ")")

        # Monta a string de formato e os argumentos para printf
        formatted_print = 'printf("' + ' '.join(format_strings) + '\\n", ' + ', '.join(print_arguments) + ');'
        self.ir.append(formatted_print)

    def control_structure(self):
        """
        control_structure ::= if ( expr ) { instructions } [ elif ( expr ) { instructions } ]* [ else { instructions } ]
        """
        print("Parsing control structure...")
        found_if = False  # Estado para verificar se um 'if' foi encontrado

        # Processa o if
        if self.token_atual.tipo == T_IF:

            self.if_Scope = Scope(self.currentScope)
            self.currentScope = self.if_Scope

            found_if = True
            self.use(T_IF)
            self.use(T_PARA, "(")
            condition, condition_str = self.expr()
            self.use(T_PARA, ")")
            self.use(T_BLOCK, "{")
            self.ir.append(f"if ({condition_str}) {{")
            self.instructions()
            self.use(T_BLOCK, "}")
            self.ir.append("}")

            self.currentScope = self.currentScope.parent

        # Processa um ou mais elifs
        while self.token_atual.tipo == T_ELIF:
            if not found_if:
                raise Exception("Erro de sintaxe: 'elif' sem um 'if' anterior.")

            self.elif_Scope = Scope(self.currentScope)
            self.currentScope = self.elif_Scope

            self.use(T_ELIF)
            self.use(T_PARA, "(")
            condition, condition_str = self.expr()
            self.use(T_PARA, ")")
            self.use(T_BLOCK, "{")
            self.ir.append(f"else if ({condition_str}) {{")
            self.instructions()
            self.use(T_BLOCK, "}")
            self.ir.append("}")

            self.currentScope = self.currentScope.parent

        # Processa o else, se existir
        if self.token_atual.tipo == T_ELSE:
            if not found_if:
                raise Exception("Erro de sintaxe: 'else' sem um 'if' anterior.")

            self.else_Scope = Scope(self.currentScope)
            self.currentScope = self.else_Scope

            self.use(T_ELSE)
            self.use(T_BLOCK, "{")
            self.ir.append("else {")
            self.instructions()
            self.use(T_BLOCK, "}")
            self.ir.append("}")

            self.currentScope = self.currentScope.parent

    def compare_expr(self):
        """
        compare_expr ::= expr <op_comp> expr
        """
        left_value, left_str = self.arith_expr()
        op = self.token_atual.valor
        self.use(T_OP_COMP)
        right_value, right_str = self.arith_expr()
        return f"{left_str} {op} {right_str}"


    def expr(self):
        """
        expr ::= arith_expr | compare_expr
        """
        print("Parsing expression...")
        if self.token_atual.tipo in [T_FLOAT, T_ID]:
            return self.arith_expr()
        elif self.token_atual.tipo == T_OP_COMP:
            return self.compare_expr()
        else:
            self.erro("Invalid expression")

    def arith_expr(self):
        """
        arith_expr ::= term expr_e_line
        """
        print("Parsing arithmetic expression...")
        term_value, term_str = self.term()
        expr_line_value, expr_line_str = self.expr_e_line(term_value)
        return expr_line_value, f"{term_str} {expr_line_str}"

    def term(self):
        """
        term ::= factor expr_t_line
        """
        print("Parsing term...")
        factor_value, factor_str = self.factor()  # ComeÃ§a com um fator
        term_line_value, term_line_str = self.expr_t_line(factor_value)  # Depois processa a linha de termos
        return term_line_value, f"{factor_str} {term_line_str}"

    def factor(self):
        """
        factor ::= ( expr ) | <id> | <int> | <float>
        """
        if self.token_atual.tipo == T_PARA and self.token_atual.valor == "(":
            self.use(T_PARA, "(")
            _, expr_str = self.expr()
            self.use(T_PARA, ")")
            return _, f"({expr_str})"
        elif self.token_atual.tipo == T_ID:
            var_name = self.token_atual.valor
            self.use(T_ID)
            if not self.currentScope.doesSymbolExist(var_name):
                raise Exception(f"Variable {var_name} not declared")

            print("Escopo atual: ", self.currentScope.doesSymbolExist(var_name).symbols)
            return self.currentScope.doesSymbolExist(var_name).symbols[var_name], var_name
        elif self.token_atual.tipo == T_FLOAT:
            literal_value = float(self.token_atual.valor)
            self.use(T_FLOAT)
            return literal_value, str(literal_value)
        else:
            self.erro("Expected a factor")

    def expr_t(self) -> int:
        """
        expr_t ::= expr_f expr_t_line
        """
        print("Parsing expression t...")
        r, s = self.expr_f()
        r2, s2 = self.expr_t_line(r)
        return r2, f"{s} {s2}"

    def expr_e_line(self, inherited_t: int):
        """
        expr_e_line ::= <op +> expr_t expr_e_line |
                        <op -> expr_t expr_e_line | epsilon
        """
        print("Parsing expression e line...")
        if self.token_atual.tipo == T_OP:
            if self.token_atual.valor == "+":
                self.use(T_OP, "+")
                a, s = self.expr_t()
                res = a + inherited_t
                a2, s2 = self.expr_e_line(res)
                str_expr = f"+ {s} {s2}"
                return a2, str_expr
            elif self.token_atual.valor == "-":
                self.use(T_OP, "-")
                a, s = self.expr_t()
                res = inherited_t - a
                a2, s2 = self.expr_e_line(res)
                str_expr = f"- {s} {s2}"
                return a2, str_expr
            elif self.token_atual.valor == "*":
                self.use(T_OP, "*")
                a, s = self.expr_t()
                res = a * inherited_t
                a2, s2 = self.expr_e_line(res)
                str_expr = f"* {s} {s2}"
                return a2, str_expr
            elif self.token_atual.valor == "/":
                self.use(T_OP, "/")
                a, s = self.expr_t()
                if a == 0:
                    raise Exception("Erro semantico. DivisÃ£o por zero.")
                res = inherited_t / a
                a2, s2 = self.expr_e_line(res)
                str_expr = f"/ {s} {s2}"
                return a2, str_expr
        elif self.token_atual.valor in ["==", "!=", "<", ">", "<=", ">="]:
            teste = self.token_atual
            self.use(T_OP_COMP)
            a, s = self.expr_t()
            res = inherited_t
            a2, s2 = self.expr_e_line(res)
            str_expr = f"/ {s} {s2}"
            return a2, str_expr

        # Prod vazia
        return 0 + inherited_t, ""

    def expr_t_line(self, inherited_t: int):
        """
        expr_t_line ::= <op *> expr_f expr_t_line |
                        <op /> expr_f expr_t_line | epsilon
        """
        print("Parsing expression t line...")
        if self.token_atual.tipo == T_OP:
            if self.token_atual.valor == "*":
                self.use(T_OP, "*")
                a, s = self.expr_f()
                res = a * inherited_t
                a2, s2 = self.expr_t_line(res)
                str_expr = f"* {s} {s2}"
                return a2, str_expr
            elif self.token_atual.valor == "/":
                self.use(T_OP, "/")
                a, s = self.expr_f()
                if a == 0:
                    raise Exception("Erro semantico. DivisÃ£o por zero.")
                res = inherited_t / a
                a2, s2 = self.expr_t_line(res)
                str_expr = f"/ {s} {s2}"
                return a2, str_expr
        elif self.token_atual.tipo == T_OP_COMP:
            comp = self.token_atual
            self.use(T_OP_COMP)
            a, s = self.expr_f()
            res = inherited_t
            a2, s2 = self.expr_t_line(res)
            str_expr = f"{comp.valor} {s} {s2}"
            return a2, str_expr
        # Prod vazia
        return 1 * inherited_t, ""

    def expr_f(self):
        """
        expr_f ::= ( expr ) | <id> | <float> | sqrt ( expr ) | <int> ^ <int>
        """
        print("Parsing expression f...")
        str_expr = ""
        if self.token_atual.tipo == T_PARA:
            self.use(T_PARA, "(")
            res, s = self.expr()
            self.use(T_PARA, ")")
            str_expr = f"({s})"
        elif self.token_atual.tipo == T_ID:
            if not self.currentScope.doesSymbolExist(self.token_atual.valor):
                raise Exception(f"Erro Semantico, variavel {self.token_atual.valor} nao foi declarada")
            res = self.currentScope.doesSymbolExist(self.token_atual.valor).symbols #self.currentScope.symbols[self.token_atual.valor]
            str_expr = self.token_atual.valor
            self.use(T_ID)
        elif self.token_atual.tipo == T_FLOAT:
            str_expr = str(self.token_atual.valor)
            res = float(self.token_atual.valor)
            self.use(T_FLOAT)
        else:
            self.erro()

        return res, str_expr

##############################################################################

arquivo = open('codigo.x','r')
ln = 1

tokens = []

for l in arquivo.readlines():

    # analisador lexico

    l = l.replace('\n','')

    for token in l.split():
        try:
            tokens.append(afd_principal(token))
        except Exception as e:
            print(tokens)
            print(str(e) + " na posiÃ§Ã£o %i da linha %i - %s" % (l.index(token), ln, token))
            raise StopExecution
    ln += 1

print([str(t) for t in tokens])

# analisador sintatico

parser = Parser(tokens)
parser.start()
code = parser.generate_ir()

with open("codigo.c", "w") as out_f:
    out_f.write(code)
```
